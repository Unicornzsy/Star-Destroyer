```python
s = list(np.linspace(0, 8, 100))
slice = list(np.linspace(0, 8, 9))
ls, ls2, yerr = [], [], []
for i in range(100):
    ls.append(eb(s[i]))
plt.plot(s, ls,  'r--', lw = 3, label = r'$ideal(c = 0.205)$')

for i in range(9):
    ls2.append(proportion(sumb[i]))
    yerr.append(eb(slice[i]) - proportion(sumb[i]))
plt.scatter(slice, ls2, s = 15, c = 'k')
plt.errorbar(slice, ls2, yerr,capsize=3, ecolor='k', label = r'$real$')

xlabel(r'$x / slice$', fontsize = 14)
ylabel(r'$e^{-x\cdot c}$', fontsize = 14)
plt.grid()
plt.legend(frameon = False)
plt.show()
```
