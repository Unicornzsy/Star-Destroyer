```python
from scipy import interpolate
```

## 实验十

```python
f_o = [ log(52.1688), log(500), log(700), log(1000), log(2500), log(4000), log(5000), log(9060.08)]#
A_o = [ 854.15, 1230.2, 1232, 1234.30, 1232, 1231, 1230.8, 861.02]#
f_c = [ log(15.2394), log(400), log(1000), log(10000), log(22986.4)]
A_c = [ 53.1155, 73.06, 73.9525, 73.13, 54.153]
```

```python
plt.plot(f_o, A_o, c = 'r', label = r'$open loop$')
plt.plot(f_c, A_c, c = 'b', label = r'$closed loop$')

xlabel(r'$log(Hz)$', fontsize = 14)
ylabel(r'$A$', fontsize = 14)
plt.grid()
plt.legend(frameon = False)
plt.show()
```

```python
xo = np.linspace(log(40), log(10000), 100000)
qo = interpolate.splrep(f_o, A_o, k=4, s=0.5)
fo = interpolate.splev(xo , qo)

plt.plot(xo, fo, c = 'r', label = r'$open \ loop$')

xc = np.linspace(log(10), log(24000), 100000)
qc = interpolate.splrep(f_c, A_c, k=3, s=0.1)
fc = interpolate.splev(xc , qc)


plt.plot(xc, fc, c = 'b', label = r'$closed \ loop$')

plt.hlines(1234.4*0.707, 2, 12, lw = 1, ls = '--')
plt.hlines(73.9525*0.707, 2, 12, lw = 1, ls = '--')
xlim(2, 12)
xlabel(r'$log(f)$', fontsize = 14)
ylabel(r'$A$', fontsize = 14)
plt.grid()
plt.legend(frameon = False)
plt.show()
```

## 实验九

```python
fi = [ log(63), log(500), log(700), log(1000), log(1200), log(2000), log(2500), log(3750), log(5368)]#
A = [ 545.630, 770.8, 771, 772, 771.8, 771, 770.6, 769.8, 545.792]#
```

```python
x = np.linspace(log(58), log(5500), 100000)
q = interpolate.splrep(fi, A, k=5, s=0.2)
f = interpolate.splev(x , q)

plt.plot(x, f, c = 'r')#, label = r'$open \ loop$'

plt.hlines(772*0.707, 2, 12, lw = 1, ls = '--')

xlim(3.5, 9)
xlabel(r'$log(f)$', fontsize = 14)
ylabel(r'$A$', fontsize = 14)
plt.grid()
#plt.legend(frameon = False)
plt.show()
```

```python
f_i1 = [log(60), log(80), log(150), log(300), log(500), log(800), log(1000), log(1500), log(2000), log(4000), log(5000)]
A_1 = [489.7, 552.3, 642.8, 684.9, 700.1, 693.1, 681.4, 673.1, 651.1, 541.9, 489.7]
x_1 = np.linspace(log(50), log(5200), 100000)
q_1 = interpolate.splrep(f_i1, A_1, k=3, s=1.2)
f_1 = interpolate.splev(x_1 , q_1)

plt.scatter(f_i1, A_1, s = 10, c = 'k')
plt.plot(x_1, f_1, c = 'r', label = r'$real$')

plt.hlines(700.1*0.707, 2, 12, lw = 1, ls = '--', colors = 'orange')

plt.plot(x, f, c = 'b', label = r'$ideal$')
plt.hlines(772*0.707, 2, 12, lw = 1, ls = '--', colors = 'cyan')

xlim(3.5, 9)
xlabel(r'$log(f)$', fontsize = 14)
ylabel(r'$A$', fontsize = 14)
plt.grid()
plt.legend(frameon = False)
plt.show()
```
