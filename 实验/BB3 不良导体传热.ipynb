## 有机玻璃测温仪

```python
t1 = [28.1, 28.8, 29.5, 30.2, 30.9, 31.6, 32.3, 32.9, 33.5, 34.1, 34.7, 35.4, 36.0]
t2 = [34.0, 34.7, 35.4, 36.2, 36.9, 37.6, 38.3, 38.9, 39.5, 40.1, 40.7, 41.4, 42.0]
t = np.linspace(8, 20, 13)
poly1 = np.polyfit(t, t1, 1)
plt.plot(t, np.polyval(poly1, t), label = '$T_1$ {}'.format(np.poly1d(poly1)))
plt.scatter(t, t1, c = 'k')

poly2 = np.polyfit(t, t2, 1)
plt.plot(t, np.polyval(poly2, t), label = '$T_2$ {}'.format(np.poly1d(poly2)))
plt.scatter(t, t2, c = 'k')

plt.xlabel('t/min', fontsize = 14)
plt.ylabel('$T/{}^{\circ}C$', fontsize = 14)
plt.legend(frameon = False)
plt.show()
```
