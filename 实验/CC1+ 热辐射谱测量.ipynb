### 普朗克黑体辐射定律

```python
def planck(l, t):
    return ((3.7415e-16)/((l*1e-9)**5)/(e**(1.4388e-2/((l*1e-9)*t))-1))*1e-9
```

```python
p_2999, p_2674, p_2455, p_2208 = [], [], [], []
for i in range(1701):
    p_2999.append(planck(datas_t['lambda'][i], 2999))
    p_2674.append(planck(datas_t['lambda'][i], 2674))
    p_2455.append(planck(datas_t['lambda'][i], 2455))
    p_2208.append(planck(datas_t['lambda'][i], 2208))
```

```python
plt.subplots(figsize = (16,9))
plt.plot(datas_t['lambda'][:], datas_t['energy_2999'][:], c = 'k', label = r'$2999\kappa$')
plt.plot(datas_t['lambda'][:], p_2999, 'r--', lw = 1)

plt.plot(datas_t['lambda'][:], datas_t['energy_2674'][:], c = 'magenta', label = r'$2674\kappa$')
plt.plot(datas_t['lambda'][:], p_2674, 'r--', lw = 1)

plt.plot(datas_t['lambda'][:], datas_t['energy_2455'][:], c = 'yellow', label = r'$2455\kappa$')
plt.plot(datas_t['lambda'][:], p_2455, 'r--', lw = 1)

plt.plot(datas_t['lambda'][:], datas_t['energy_2208'][:], c = 'cyan', label = r'$2208\kappa$')
plt.plot(datas_t['lambda'][:], p_2208, 'r--', lw = 1)

title(r'$Planck\'s \ Law$')
xlabel(r'$\lambda/nm$', fontsize = 14)
ylabel(r'$P/(W/mm^3)$', fontsize = 14)
plt.legend(frameon = False)
plt.grid()
plt.show()
```
