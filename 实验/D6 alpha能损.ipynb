测量时间的比较

```python
df_1003 = pd.read_csv(r'C:\Users\31260\Desktop\近代物理实验\D6 alpha原子\D6\100v(3s).txt')
dt_1003 = df_1003.head(1030)
df_10030 = pd.read_csv(r'C:\Users\31260\Desktop\近代物理实验\D6 alpha原子\D6\100v(30s).txt')
dt_10030 = df_10030.head(1030)
df_100300 = pd.read_csv(r'C:\Users\31260\Desktop\近代物理实验\D6 alpha原子\D6\100v(300s).txt')
dt_100300 = df_100300.head(1030)
```

```python
x = np.linspace(1, 1024, 1023)

plt.subplots(figsize = (16, 9))
plt.plot(x, dt_1003*23, label = r'$T_{detect} = 3s(\times 23)$')
plt.plot(x, dt_10030*3.8, label = r'$T_{detect} = 30s(\times 3.8)$')
plt.plot(x, dt_100, label = r'$T_{detect} = 120s$')
plt.plot(x, dt_100300/2.7, label = r'$T_{detect} = 300s(\times \frac{1}{2.7})$')

plt.xlabel(r'$Channel\quad number$', fontsize = 14)
plt.ylabel(r'$atom(Count)$', fontsize = 14)
plt.xlim(0, 1024)
plt.ylim(-10, 350)
plt.grid()
plt.legend(frameon = False)
plt.show()
```

```python
fig, ax = plt.subplots(1, 4, figsize = (14, 4))

#相对于120s做差比较
dtm_1003 = np.array(dt_1003*23) - np.array(dt_100)
dtm_10030 = np.array(dt_10030*3.8) - np.array(dt_100)
dtm_100 = np.array(dt_100) - np.array(dt_100)
dtm_100300 = np.array(dt_100300/2.7) - np.array(dt_100)

ax[0].plot(x, dtm_1003, c = 'cyan', label = r'$T_{detect} = 3s(\times 23)$'); ax[0].set_xlim(0, 1024); ax[0].set_ylim(-110, 100)
ax[0].grid(); ax[0].legend(frameon = False)
ax[0].set_xlabel(r'$Channel\quad number$', fontsize = 10)
ax[0].set_ylabel(r'$atom(Count)$', fontsize = 10)

ax[1].plot(x, dtm_10030, c = 'magenta', label = r'$T_{detect} = 30s(\times 3.8)$'); ax[1].set_xlim(0, 1024); ax[1].set_ylim(-110, 100)
ax[1].grid(); ax[1].legend(frameon = False)
ax[1].set_xlabel(r'$Channel\quad number$', fontsize = 10)

ax[2].plot(x, dtm_100, c = 'yellow', label = r'$T_{detect} = 120s$'); ax[2].set_xlim(0, 1024); ax[2].set_ylim(-110, 100)
ax[2].grid(); ax[2].legend(frameon = False)
ax[2].set_xlabel(r'$Channel\quad number$', fontsize = 10)

ax[3].plot(x, dtm_100300, c = 'k', label = r'$T_{detect} = 300s(\times \frac{1}{2.7})$'); ax[3].set_xlim(0, 1024); ax[3].set_ylim(-110, 100)
ax[3].grid(); ax[3].legend(frameon = False)
ax[3].set_xlabel(r'$Channel\quad number$', fontsize = 10)


plt.show()
```

```python
x = [792.89, 846.48]
E = [5.155, 5.486]


a = (E[0]-E[1])/(x[0]-x[1])
b = E[0] - a*x[0]
f = lambda x: a*x + b

x_dense = np.linspace(700, 900, 1000)
plt.scatter(x, E, c = 'k')
plt.plot(x_dense, f(x_dense), label = r'$E = {:.3f}x + {:.3f}$'.format(a, b))

plt.xlabel(r'$Channel\quad number$', fontsize = 14)
plt.ylabel(r'$Energy/MeV$', fontsize = 14)
plt.grid()
plt.legend(frameon = False)
plt.show()
```
