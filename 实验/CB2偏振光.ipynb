```python
def malus(theta):
   # if  ((np.cos(theta))**2) <= 0.5:
    return (np.cos(theta/180*np.pi))**2
    #elif ((np.cos(theta))**2) > 0.5:
        #return (np.cos(theta))**2-0.5
```

```python
x0 = np.linspace(0,180,10)
P = [17.08e-6, 19.03e-6, 18.21e-6, 14.54e-6, 9.87e-6, 6.76e-6, 6.30e-6, 8.93e-6, 13.20e-6, 17.43e-6]
```

## 自然光

```python
plt.scatter(malus(x0), P, c = 'k')
plt.plot(malus(x0), P, c = 'k')

ylim(0,0.00002)
title(r'$natural \ light$')
xlabel(r'$cos^2\theta$', fontsize = 14)
ylabel(r'$P/W$', fontsize = 14 )
plt.grid()
plt.show()
```

```python
plt.plot(x0, P, c = 'k')

title(r'$natural \ light$')
xlabel(r'$\theta/{}^{\circ}$', fontsize = 14)
ylabel(r'$P/W$', fontsize = 14 )
plt.grid()
plt.show()
```

## 绿色偏振片

```python
P_green = [0.040e-6, 0.048e-6, 0.052e-6, 0.038e-6, 0.028e-6, 0.018e-6, 0.019e-6, 0.017e-6, 0.030e-6, 0.037e-6]
```

```python
plt.scatter(malus(x0), P_green, c = 'g')
plt.plot(malus(x0), P_green, c = 'g')

title(r'$green \ light$')
ylim(0,0.00000006)
xlabel(r'$cos^2\theta$', fontsize = 14)
ylabel(r'$P/W$', fontsize = 14 )
plt.grid()
plt.show()
```

```python
plt.plot(x0, P_green, c = 'g')

title(r'$green \ light$')
xlabel(r'$\theta/{}^{\circ}$', fontsize = 14)
ylabel(r'$P/W$', fontsize = 14 )
plt.grid()
plt.show()
```

```python
P_green = [0.040e-6, 0.048e-6, 0.052e-6, 0.038e-6, 0.028e-6, 0.018e-6, 0.019e-6, 0.017e-6, 0.030e-6, 0.037e-6]
for i in range(len(P_green)):
    P_green[i] = P_green[i]*370
```

```python
P_red = [6.60e-6, 7.37e-6, 6.93e-6, 5.50e-6, 3.86e-6, 2.66e-6, 2.49e-6, 3.39e-6, 5.09e-6, 5.66e-6]
for i in range(len(P_red)):
    P_red[i] = P_red[i]*2.6
```

```python
plt.scatter(malus(x0), P, c = 'k', label = r'$natural \ light$')
plt.plot(malus(x0), P, c = 'k')

plt.scatter(malus(x0), P_red, c = 'r', label = r'$red \ light(\times 2.6)$')
plt.plot(malus(x0), P_red, c = 'r')

plt.scatter(malus(x0), P_green, c = 'g', label = r'$green \ light(\times 370)$')
plt.plot(malus(x0), P_green, c = 'g')

title(r'$light \ compared$')
ylim(0,0.00002)
xlabel(r'$cos^2\theta$', fontsize = 14)
ylabel(r'$P/W$', fontsize = 14 )
plt.legend(frameon = False)
plt.grid()
plt.show()
```
