```python
x = [4, 5, 6, 7]
fx = [12.46, 12.97, 13.11, 13.48]
fw = [12.46, 12.89, 13.11, 13.48]

plt.plot(x, fx, 'b-', label =  'object focal length')
plt.scatter(x, fx, c = 'k', s = 20)
plt.xlabel(r'$\delta x/cm$', fontsize = 14)
plt.ylabel(r'$f/cm$', fontsize = 14)

plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot( x, fw, 'orange', label =  'image focal length')
plt.scatter( x, fw, c='k', s = 20)

plt.xlabel(r'$\delta x/cm$', fontsize = 14)
plt.ylabel(r'$f^{,}/cm$', fontsize = 14)

plt.legend()
plt.grid()
plt.show()
```
