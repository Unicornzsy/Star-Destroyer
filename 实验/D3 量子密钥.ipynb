```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

波片

```python
def ML(theta, i, d):
    return i*(np.cos(theta/180*2*np.pi+ d/180*np.pi))**2 

def P(p):
    return 10**(p/10)
```

```python
P(9.01)
```

```python
theta = np.linspace(0, 360, 17)
p1 = [9.01, -2.30, 4.78, 10.32, 8.98, -3.13, 5.25, 10.35, 8.96, -3.08, 4.85, 10.41, 9.03, -2.43, 5.19, 10.28, 9.23]
P1_t = [P(p1[x]) for x in range(len(p1))]

p2 = [6.90, 10.62, 8.23, -10.28, 6.74, 10.56, 8.23, -7.80, 6.98, 10.50, 8.11, -12.69, 7.02, 10.63, 8.14, -7.19, 6.98]
P2_t = [P(p2[x]) for x in range(len(p2))]
```

```python
popt, pocv = curve_fit(ML, theta, P1_t)

theta_dense = np.linspace(0, 360, 1000)

plt.subplots(figsize = (16, 9))
#plt.plot(theta, P1_t)
plt.plot(theta_dense, ML(theta_dense, popt[0], popt[1]), c = 'orange', label = r'regression')
plt.scatter(theta, P1_t, c = 'b', label = 'measurement')
t = 163.5
plt.text(166, 11.2, r'$maxium(\theta=163.5^{\circ})$', fontsize = 14, c = 'k')
plt.scatter(t, ML(t, popt[0], popt[1]), c = 'k')

plt.ylabel(r'$P/mW$', fontsize = 14)
plt.xlabel(r'$\theta/^{\circ}$', fontsize = 14)
plt.legend(frameon = False)
plt.grid()
plt.show()
```

```python
popt, pocv = curve_fit(ML, theta, P2_t)

theta_dense = np.linspace(0, 360, 1000)

plt.subplots(figsize = (16, 9))

plt.plot(theta_dense, ML(theta_dense, popt[0], popt[1]), c = 'orange', label = r'regression')
plt.scatter(theta, P2_t, c = 'b', label = 'measurement')
t = 160.0
plt.text(166, 0, r'$minium(\theta=160.0^{\circ})$', fontsize = 14, c = 'k')
plt.scatter(t, ML(t, popt[0], popt[1]), c = 'k')

plt.ylabel(r'$P/mW$', fontsize = 14)
plt.xlabel(r'$\theta/^{\circ}$', fontsize = 14)
plt.legend(frameon = False)
plt.grid()
plt.show()
```
