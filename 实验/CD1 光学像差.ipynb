```python
x_p = [0, 0.78,1.17]
theta = [0, 30, 45]
```

```python
#plt.plot(x_p, theta)
plt.scatter(x_p, theta, c = 'k', s = 15)
q = np.polyfit(x_p,theta,1);
plt.plot(x_p,np.polyval(q,x_p), label = r'$\theta = {:.3f}x_p+{:.3f}$'.format(q[0], q[1]))

xlabel(r'$x_p/cm$', fontsize = 14)
ylabel(r'$\theta/{}^{\circ}$', fontsize = 14)
plt.legend(frameon = False)
plt.grid()
plt.show()
```
