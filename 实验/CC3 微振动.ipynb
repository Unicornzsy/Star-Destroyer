```python
def wave(n_1, n_2, a, b):
    return (n_1 + n_2 + np.arcsin(a/np.pi/2) + np.arcsin(b/np.pi/2))
```

```python
def wave_A(n_1, n_2, a, b):
    return (n_1 + n_2 + np.arcsin(a/np.pi/2) + np.arcsin(b/np.pi/2))/200
```

```python
n1_0 = [13, 13, 14]
n2_0 = [1.25, 0.25, 0.5]
a_0 = [0.20, 0, 0]
b_0 = [0, 0.15, 0.05]
q_0 = []
for i in range(3):
    q_0.append(wave(n1_0[i], n2_0[i], a_0[i], b_0[i]))
q_0
```

```python
n1 = [5 ,7 ,8, 10, 13, 9, 7, 6, 5]
n2 = [1, 0.25, 0.75, 0, 0.25, 0.25, 0.5, 1, 1, 1]
a = [0.46, 0.08, 0.30, 0.10, 0.02, 0, 0.05, 0.12, 0.20]
b = [0.18, 0.22, 0.10, 0, 0.01, 0, 0.05, 0.03, 0.05]
q = []
for i in range(9):
    q.append(wave_A(n1[i], n2[i], a[i], b[i]))
```
