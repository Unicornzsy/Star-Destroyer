# 杂波分析

```python
from scipy import fftpack
```

```python
plt.plot(DATA2['t'][24300:24500], DATA2['Hv'][24300:24500], c='orange', label='H')

plt.grid()
plt.legend(frameon=False)

plt.xlabel(r'$t/s$', fontsize=14)
plt.ylabel(r'$U/V$', fontsize=14)

plt.show()
```

```python
t = np.linspace(0, 1, 1000)

z = 4*np.sin(2*np.pi*200*t) + 2*np.sin(2*np.pi*400*t )  

plt.plot(t, z, )

plt.xlim(0,0.1)
print('差倍数频率的波有相似形状，猜测可能是电路中有多个杂波合成')
```

```python
y = fftpack.fft(z)
plt.plot(y)

print("由网上资料得知，后部对称主要为虚部影响")
plt.show()
```

```python
a = np.linspace(0, 1, 20)
b = np.sin(2*np.pi*a)
y = fftpack.fft(b)
plt.plot(y)

plt.xlabel('f/Hz', fontsize=14)
plt.ylabel('U/V', fontsize=14)
print('由此简单分解可知,横轴表示对应频率，纵轴表示对应频率的振幅')
plt.show()
```

```python
c = fftpack.ifft(y)
plt.plot(a, c, lw=5)
plt.plot(a, b, c='k', alpha=0.4)
plt.show()
```
